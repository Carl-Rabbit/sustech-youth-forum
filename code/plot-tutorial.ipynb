# 点图

Matplotlib提供了绘制点图的方法（scatter），在这个方法中，第一个参数是x，第二个参数是y。
调用这个方法，我们很容易就能绘制出最基本的点图。

````python
import matplotlib.pyplot as plt

plt.scatter(df.index, df['newConfirm'])
````

# 多图

Matplotlib不仅能一次性画一张图，还能在同一张图中包含多个子图。

首先我们生成一组用于绘制多图的数据。

````python
x_labels = ['04.05', '04.10', '04.15']
x_label_indexes = [0, 5, 10]

x1 = df.index[:11]
y1 = df['newConfirm'][:11]

x2 = df.index[:11]
y2 = df['newRecovered'][:11]

x3 = df.index[:11]
y3 = df['newConfirm'][:11] * 10

x4 = df.index[:11]
y4 = df['newRecovered'][:11] * 10

x5 = df.index[:11]
y5 = df['newConfirm'][:11] * 100

x6 = df.index[:11]
y6 = df['newRecovered'][:11] * 100
````

绘制一个基本的包含子图的图

```python3
matplotlib.pyplot.subplot(*args, **kwargs)
```
[官方文档](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.subplot.html)

````python
fig = plt.figure(figsize=[10, 6])

width = 0.4

ax = fig.add_subplot(221)

ax.bar(x1 - width / 2, y1, width, label='Label A', hatch='...', color='coral', edgecolor='black')
ax.bar(x2 + width / 2, y2, width, label='Label B', hatch='///', color='skyblue', edgecolor='black')
ax.set_xlabel('X label', fontsize=14)
ax.set_ylabel('Y label', fontsize=14)
ax.set_xticks(x_label_indexes, x_labels, fontsize=14)
ax.set_yticks([0, 1000, 2000, 3000, 4000, 5000, 6000], ['0', '1k', '2k', '3k', '4k', '5k', '6k'], fontsize=14)
ax.set_xlim(-1, 11)
ax.tick_params(axis='both', direction='in')
ax.tick_params(axis='x', size=0)
ax.legend()

ax = fig.add_subplot(222)

ax.bar(x3 - width / 2, y3, width, label='Label A', hatch='...', color='coral', edgecolor='black')
ax.bar(x4 + width / 2, y4, width, label='Label B', hatch='///', color='skyblue', edgecolor='black')
ax.set_xlabel('X label', fontsize=14)
ax.set_ylabel('Y label', fontsize=14)
ax.set_xticks(x_label_indexes, x_labels, fontsize=14)
ax.set_yticks([0, 10000, 20000, 30000, 40000, 50000, 60000], ['0', '10k', '20k', '30k', '40k', '50k', '60k'], fontsize=14)
ax.set_xlim(-1, 11)
ax.tick_params(axis='both', direction='in')
ax.tick_params(axis='x', size=0)
ax.legend()

ax = fig.add_subplot(212)

ax.bar(x5 - width / 2, y5, width, label='Label A', hatch='...', color='coral', edgecolor='black')
ax.bar(x6 + width / 2, y6, width, label='Label B', hatch='///', color='skyblue', edgecolor='black')
ax.set_xlabel('X label', fontsize=14)
ax.set_ylabel('Y label', fontsize=14)
ax.set_xticks(x_label_indexes, x_labels, fontsize=14)
ax.set_yticks([0, 100000, 200000, 300000, 400000, 500000, 600000], ['0', '100k', '200k', '300k', '400k', '500k', '600k'], fontsize=14)
ax.set_xlim(-1, 11)
ax.tick_params(axis='both', direction='in')
ax.tick_params(axis='x', size=0)
ax.legend()

fig.tight_layout()
````

为每个子图添加标题
